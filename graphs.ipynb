```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import statsmodels.api as sm

fig, ax = plt.subplots(figsize=(12,8))

# "ЧАСТЬ РЕЧИ" для людей
x_speech = df["доля_хороших_предсказаний_люди"]
y_speech = df["часть речи ош. люди"]


df_points_speech = pd.DataFrame({'x': x_speech, 'y': y_speech})
counts_speech = df_points_speech.groupby(['x', 'y']).size().reset_index(name='counts')
sizes_speech = 30 + (counts_speech['counts'] - 1) * 40  # Масштабируем размеры точек

# Регрессия через statsmodels , построения доверительного интервала
X_speech = sm.add_constant(x_speech)
model_speech = sm.OLS(y_speech, X_speech).fit()

# Линия регрессии и доверительный интервал
x_pred_speech = np.linspace(x_speech.min(), x_speech.max(), 100)
X_pred_speech = sm.add_constant(x_pred_speech)
predictions_speech = model_speech.get_prediction(X_pred_speech)
pred_summary_frame_speech = predictions_speech.summary_frame(alpha=0.05)  # 95% доверительный интервал


ax.plot(x_pred_speech, pred_summary_frame_speech['mean'], color='blue', label='Ошибки в части речи')

ax.fill_between(x_pred_speech,
                pred_summary_frame_speech['mean_ci_lower'],
                pred_summary_frame_speech['mean_ci_upper'],
                color='blue', alpha=0.2)

ax.scatter(counts_speech['x'], counts_speech['y'], s=sizes_speech, alpha=0.8, color='blue')

# "ЛЕКСИЧЕСКИЕ ОШИБКИ"  люди
x_lexical = df["доля_хороших_предсказаний_люди"]
y_lexical = df["лекс. ошибка люди"]

df_points_lexical = pd.DataFrame({'x': x_lexical, 'y': y_lexical})
counts_lexical = df_points_lexical.groupby(['x', 'y']).size().reset_index(name='counts')
sizes_lexical = 30 + (counts_lexical['counts'] - 1) * 40

X_lexical = sm.add_constant(x_lexical)
model_lexical = sm.OLS(y_lexical, X_lexical).fit()

x_pred_lexical = np.linspace(x_lexical.min(), x_lexical.max(), 100)
X_pred_lexical = sm.add_constant(x_pred_lexical)
predictions_lexical = model_lexical.get_prediction(X_pred_lexical)
pred_summary_frame_lexical = predictions_lexical.summary_frame(alpha=0.05)

ax.plot(x_pred_lexical, pred_summary_frame_lexical['mean'], color='red', label='Лекс. ошибки')
ax.fill_between(x_pred_lexical,
                pred_summary_frame_lexical['mean_ci_lower'],
                pred_summary_frame_lexical['mean_ci_upper'],
                color='red', alpha=0.2)

ax.scatter(counts_lexical['x'], counts_lexical['y'], s=sizes_lexical, alpha=0.8, color='red')

ax.invert_xaxis()

ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")

ax.set_title("Люди: Сравнение ошибок в части речи и лексических ошибок")
ax.set_xlabel("Доля хороших предсказаний")
ax.set_ylabel("Доля ошибок")

ax.grid(True)
ax.legend()

plt.show()
```
